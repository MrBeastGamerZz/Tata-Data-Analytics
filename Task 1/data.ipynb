```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("OnlineRetail.xlsx")
print(df.head())
```

```python
# 1. Create Total Revenue column
df['Revenue'] = df['Quantity'] * df['UnitPrice']

# 2. Check for negative quantities (Canceled orders)
negative_qty = df[df['Quantity'] < 0]
print(f"Total rows with negative quantities: {len(negative_qty)}")

# 3. Basic Stats
total_revenue = df['Revenue'].sum()
print(f"Total Revenue: ${total_revenue:,.2f}")
```

```python
# Grouping by Country to see who spends the most
country_stats = df.groupby('Country')['Revenue'].sum().sort_values(ascending=False).head(38)
print(country_stats)
```

```python
country_stats = df.groupby('Country')['Revenue'].sum().sort_values(ascending=False).head(10)
print("Top 10 Countries by Revenue:")
print(country_stats)
```

```python
# Create a 'Clean' dataframe for your meeting insights
# We exclude negative quantities and rows without a CustomerID
df_clean = df[(df['Quantity'] > 0) & (df['CustomerID'].notnull())]

print(f"Rows after cleaning: {len(df_clean)}")
print(f"Percentage of data kept: {round(len(df_clean)/len(df) * 100, 2)}%")
```
